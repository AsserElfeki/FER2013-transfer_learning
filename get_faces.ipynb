```python
import matplotlib.pyplot as plt
plt.gray()
import cv2


face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
img = cv2.imread('image.jpg')
plt.imshow(img)


```

```python
# plt.imshow(gray_img)
faces = face_cascade.detectMultiScale(
    img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
# plt.imshow(faces)
for (x, y, w, h) in faces:
    cropped_img = img[y:y+h, x:x+w]
    # replace 64 with the desired size
    resized_img = cv2.resize(cropped_img, (48, 48))
    gray_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
    # pass the resized image to your facial expression classification CNN for prediction
    plt.imshow(gray_img)
```
